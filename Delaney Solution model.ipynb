### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Load dataset

```python
df = pd.read_csv('delaney_solubility.csv')
df
```

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

### Splitting data

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 100)
X_train
```

```python
X_test
```

### Building the model

## Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

## Applying model to make predictions

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate the model

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_mse_train = mean_squared_error(y_train, y_lr_train_pred)
lr_r2_train = r2_score(y_train, y_lr_train_pred)

lr_mse_test = mean_squared_error(y_test, y_lr_test_pred)
lr_r2_test = r2_score(y_test, y_lr_test_pred)
```

```python
print('lr MSE train', lr_mse_train)
print('lr R2 score train', lr_r2_train)
print('lr MSE test', lr_mse_test)
print('lr R2 score test', lr_r2_test)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_mse_train, lr_r2_train, lr_mse_test, lr_r2_test]).T
lr_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE','Test R2']
lr_results
```

### Random Forest

## Training the model

```python
from sklearn.ensemble import RandomForestRegressor

RF = RandomForestRegressor(max_depth = 2, random_state = 100)
RF.fit(X_train, y_train)
```

```python
y_rf_train_pred = RF.predict(X_train)
y_rf_test_pred = RF.predict(X_test)
```

```python
y_rf_train_pred
```

```python
y_rf_test_pred
```

## Evaluate the model

```python
rf_mse_train = mean_squared_error(y_train, y_rf_train_pred)
rf_r2_train = r2_score(y_train, y_rf_train_pred)

rf_mse_test = mean_squared_error(y_test, y_rf_test_pred)
rf_r2_test = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF MSE train', rf_mse_train)
print('RF R2 score train', rf_r2_train)
print('RF MSE test', rf_mse_test)
print('RF R2 score test', rf_r2_test)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_mse_train, rf_r2_train, rf_mse_test, rf_r2_test]).T
rf_results.columns=['Method', 'Training MSE', 'Training R2', 'Test MSE','Test R2']
rf_results
```

```python
results = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
results
```

### Data visulatzation for predicted results

```python
plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, c="blue" ,alpha=0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_train, p(y_train), 'red')
plt.ylabel('Predict LogS')
plt.xlabel('Experimental LogS')
```
